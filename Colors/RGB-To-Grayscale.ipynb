```python
import numpy as np
import matplotlib.image  as mpimg
import matplotlib.pyplot as plt
```

```python
def rgbTogray(im):
    try:
        gray = np.zeros_like(im)
        r, g, b = im[:,:,0], im[:,:,1], im[:,:,2]
        gray = (0.2126 * r + 0.7152 * g + 0.0722 * b) * 255
        return np.uint8(gray)
    except ValueError:
        print("ERROR: RGB image needed.")
        return im
```

```python
lena = mpimg.imread('../Images/Lena.png')
lena_gray = rgbTogray(lena)
```

Printing figures to see the final results

```python
fig = plt.figure(figsize=(10,10))
a   = fig.add_subplot(1,2,1)
a.set_title('RGB')
a.set_xticks([])
a.set_yticks([])
plt.imshow(lena)
a = fig.add_subplot(1,2,2)
a.set_title('Grayscale')
a.set_xticks([])
a.set_yticks([])
plt.imshow(lena_gray, cmap='gray')
```
